```python
## Создайте в вашей таблице колонки:
# "Активных" - число активных случаев = 'Заражений' - 'Выздоровлений' - 'Смертей'
ncov_ru_j['Active'] = ncov_ru_j['Заражений'] - ncov_ru_j['Выздоровлений'] - ncov_ru_j['Смертей']

# "Активных на 100К населения"
ncov_ru_j['Active_100k'] = ncov_ru_j['Active'] * 100000 / ncov_ru_j['Регион население']

# "Заражений на 100К населения"
ncov_ru_j['Infected_100k'] = ncov_ru_j['Заражений'] * 100000 / ncov_ru_j['Регион население']

# "Смертей на 100К населения"
ncov_ru_j['Dead_100k'] = ncov_ru_j['Смертей'] * 100000 / ncov_ru_j['Регион население']

# "Заражений за день на 100К населения"
ncov_ru_j['Infected_100kpd'] = ncov_ru_j['Заражений за день'] * 100000 / ncov_ru_j['Регион население']

# "Смертей за день на 100К населения"
ncov_ru_j['Dead_100kpd'] = ncov_ru_j['Смертей за день'] * 100000 / ncov_ru_j['Регион население']
```

### 4. Летальность в федеральных округах

Рассчитайте летальность по федеральным округам по формуле: "Смертей" * 100 / "Заражений". Преобразуйте таблицу к следующему виду: округа (строки) х дата (столбцы). Оставьте только последние 14 дней и сохраните получившуюся таблицу в формате csv. Выведите, какой округ лидирует по уровню летальности, в отдельной клетке.

```python
ncov_ru_j = ncov_ru_j.reset_index()
ncov_ru_j.head(3)
```

```python
ncov_ru_j['летальность'] = ncov_ru_j['Смертей'] *100 / ncov_ru_j['Заражений']
```

```python
df = pd.pivot_table(ncov_ru_j, values = 'летальность', index = 'ФО', columns = 'Дата').iloc[:, -14:]
df
```

```python
# to determine leader for the last 14 days counnt a sum

df['Total']= df.sum(axis=1)
```

```python
# leader is; по уровню летальность за послпедние 2 недели
df[df['Total'] == max(df['Total'])].index
```

```python
df.to_csv('out.csv')
```
